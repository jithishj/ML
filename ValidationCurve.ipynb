```python
from sklearn.model_selection import validation_curve
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
digits=load_digits()
```

```python
X, y=digits.data, digits.target
param_range = np.arange(1, 11, 2)
```

```python
train_scores, test_scores=validation_curve(KNeighborsClassifier(), X, y, param_name='n_neighbors', param_range=param_range, cv=3, scoring='accuracy')
```

```python
m_train=np.mean(train_scores, axis=1)
```

```python
m_test=np.mean(test_scores, axis=1)
```

```python
fig, ax=plt.subplots()
ax.plot(param_range, m_train, label='Training')
ax.plot(param_range, m_test, label='Validation')
ax.legend()
ax.set(xlabel='Dataset size')
ax.set(ylabel='Accuracy')
fig.suptitle('Validation_Curves')
plt.show()
```
