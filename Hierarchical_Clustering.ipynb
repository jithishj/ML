```python
>>> from sklearn.datasets import make_blobs
>>> X, y = make_blobs(n_samples=800, centers=3, n_features=2,
...                   random_state=0)
```

```python
X
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, fcluster
```

```python
z=linkage(X, 'complete') #single, average, ward, centroid
```

```python
labels=fcluster(z, 3, 'maxclust')#maxclust
```

```python
labels
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```
